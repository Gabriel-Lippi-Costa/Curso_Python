## **Encontrando a amplitude de um conjunto de dados**

**Exercício:** Duas empresas contrataram 10 formandos cada. O salário inicial para cada formando é mostrado nas tabelas.

- Salários iniciais para a empresa A (em milhares de dólares):
```Python
41    38    39    45    47    41    44    41    37    42
```
- Salários iniciais para a empresa B (em milhares de dólares):
```Python
40    23    41    50    49    32    41    29    52    58
```

Encontre a amplitude dos salários iniciais para as empresas.

## Identificar os mínimos e máximos:

- Identificar os valores máximos e mínimos para cada empresa importanco cada conjunto de dados utilizando a função `Series` do Pandas
- Calcular a amplitude subtraindo o valor mínimo do valor máximo para cada conjunto de dados utilizando as funções `.max()` e `min()`

````python
# Importa a biblioteca Pandas
import pandas as pd

# Salários iniciais para as empresas (em milhares de dólares)
dados = {
    'Salários Empresa A': [41, 38, 39, 45, 47, 41, 44, 41, 37, 42],
    'Salários Empresa B': [40, 23, 41, 50, 49, 32, 41, 29, 52, 58]
}

# Criar o DataFrame
df = pd.DataFrame(dados)

# Exibindo o DataFrame
display(df)

# Calcular a amplitude para a empresa A
amplitude_a = df['Salários Empresa A'].max() - df['Salários Empresa A'].min()

# Calcular a amplitude para a empresa B
amplitude_b = df['Salários Empresa B'].max() - df['Salários Empresa B'].min()

# Exibir os resultados
print(f"  A amplitude dos salários na empresa A é: {amplitude_a} mil dólares")
print(f"  A amplitude dos salários na empresa B é: {amplitude_b} mil dólares")
````

## Média vs Mediana vs Moda

Observar como a *média*, *mediana* e *moda* se comportam para os dois conjuntos de dados:

**Em Python, recordando alguns comandos:**

````python
# Para o conjunto de dados da Empresa A
# Cálculo da média, mediana e moda
media_a = df['Salários Empresa A'].mean()
mediana_a = df['Salários Empresa A'].median()
moda_a = df['Salários Empresa A'].mode().iloc[0]

# Print com 1 casa decimal
print(f'Empresa A: média é {media_a:.1f}.')
print(f'Empresa A: mediana é {mediana_a:.1f}.')
print(f'Empresa A: moda é {moda_a:.1f}.')


# Para o conjunto de dados da Empresa B
# Cálculo da média, mediana e moda
media_b = df['Salários Empresa B'].mean()
mediana_b = df['Salários Empresa B'].median()
moda_b = df['Salários Empresa B'].mode().iloc[0]

# Print com 1 casa decimal
print(f'\nEmpresa B: média é {media_b:.1f}.')
print(f'Empresa B: mediana é {mediana_b:.1f}.')
print(f'Empresa B: moda é {moda_b:.1f}.')
````

## **Encontrando a variância <u>POPULACIONAL</u> de um conjunto de dados**

Sendo assim, calcula-se a soma dos quadrados dos desvios, SSx.

Em uma população, a média dos quadrados dos desvios é a variância populacional.



**Exercício:** 

- Salários iniciais para a empresa A (em milhares de dólares):
```Python
41    38    39    45    47    41    44    41    37    42
```

Encontre a variância e o desvio padrão populacionais dos salários iniciais para a empresa A.

**Em Python, alguns comandos novos:**

```Python
# Calcula os quadrados dos desvios parciais
df_a['Quadrados (x-u)^2'] = df_a['Desvio (x-u)']**2

# Calcular a variância populacional na Empresa X
variancia_a = df_a['Salários Empresa X'].var(ddof=0)
```

**Explicação do Código**
- `.var(ddof=0)`: O método `.var()` em Pandas calcula a variância. O parâmetro `ddof=0` (degrees of freedom) ajusta o divisor para `N`, que é usado para calcular a variância populacional. O padrão `ddof=1` é usado para calcular a variância amostral `n-1` que será visto mais à frente.

````python
# Calcula os quadrados dos desvios parciais
df_a['Quadrados (x-u)^2'] = df_a['Desvio (x-u)']**2

# Exibindo o DataFrame
display(df_a)

# Calcular a soma de todos os quadrados na Empresa A
soma_quadrados_a = df_a['Quadrados (x-u)^2'].sum()

# Calcular a variância populacional na Empresa A
variancia_a = df_a['Salários Empresa A'].var(ddof=0)

# Exibindo os quadrados totais e a variância na Empresa A
print(f'Total dos quadrados na Empresa A (SSx): {soma_quadrados_a:.1f}')
print(f'Variância populacional na Empresa A (sigma^2): {variancia_a:.1f}')
````

## **Encontrando o desvio padrão <u>POPULACIONAL</u> de um conjunto de dados**

Uma desvantagem da variância populacional $\sigma^{2}$ é que sua unidade de medida é diferente da unidade de medida do conjunto de dados.
Por exemplo, a variância para os salários iniciais no exemplo foi em “milhares de dólares quadrados”.

Para superar esse problema, tiramos a raiz quadrada da variância para obter o desvio padrã


**Exercício:**

```Python
41    38    39    45    47    41    44    41    37    42
```

**Em Python, alguns comandos novos:**

```Python
# Calcular o desvio padrão populacional na Empresa X
desvio_a = df_a['Salários Empresa X'].std(ddof=0)
```

**Explicação do Código**
- `.std(ddof=0)`: Assim como para a variância, o método `.std()` em Pandas calcula o desvio padrão. O parâmetro `ddof=0` (degrees of freedom) ajusta o divisor para `N`, que é usado para calcular o desvio padrão populacional. O padrão `ddof=1` é usado para calcular o desvio padrão amostral `n-1` que será visto mais à frente.




````python
# Calcular o desvio padrão populacional na Empresa A
desvio_a = df_a['Salários Empresa A'].std(ddof=0)

# Exibindo os quadrados totais e a variância na Empresa A
print(f'Desvio padrão populaconal na Empresa A (sigma): {desvio_a:.1f} milhares de dólares')
````

## **Encontrando a variância e o desvio padrão <u>AMOSTRAL</u>**

As expressões para a  variância amostral ($s^{2}$) e o desvio padrão amostral (s) diferem ligeiramente daquelas válidas para uma população



**Exercício:** Em um estudo com jogadores de futebol americano do ensino médio que sofreram lesões, os pesquisadores colocaram os jogadores em dois grupos:
- Jogadores que se recuperaram das concussões em 14 dias ou menos foram colocados no grupo 1.
- Aqueles que levaram mais de 14 dias foram para o grupo 2.
Os tempos de recuperação (em dias) para o grupo 1 estão listados a seguir.

```Python
4    7    6    7    9    5    8    10    9    8    7    10
```

*(Adaptado de: The American Journal of Sports Medicine.)*

Encontre a variância e o desvio padrão amostrais dos tempos de recuperação.

**Em Python, alguns comandos novos:**

```Python
# Calcular a variância amostral
variancia_amostral = dados.var(ddof=1)

# Calcular o desvio padrão amostral
desvio_padrao_amostral = dados.std(ddof=1)
```

**Explicação do Código**
- `.var(ddof=1)` e `.std(ddof=1)`: O parâmetro `ddof=1` (degrees of freedom) ajusta o divisor para `n-1`, que é usado para calcular a variância e o desvio padrão amostral.

````python
# Importa a biblioteca Pandas
import pandas as pd

# Tempos de recuperação para o grupo 1
dados = {
    'Tempo Grupo 1': [4, 7, 6, 7, 9, 5, 8, 10, 9, 8, 7, 10]
}

# Criar o DataFrame
df_grupo1 = pd.DataFrame(dados)

# Cálculo da média
media_grupo1 = df_grupo1['Tempo Grupo 1'].mean()

# Calcula os devios parciais
df_grupo1['Desvio (x-u)'] = df_grupo1['Tempo Grupo 1'] - media_grupo1

# Calcula os quadrados dos desvios parciais
df_grupo1['Quadrados (x-u)^2'] = df_grupo1['Desvio (x-u)']**2

# Exibindo o DataFrame
display(df_grupo1)


# Calcular a variância amostral
variancia_amostral = df_grupo1['Tempo Grupo 1'].var(ddof=1)

# Calcular o desvio padrão amostral
desvio_padrao_amostral = df_grupo1['Tempo Grupo 1'].std(ddof=1)

# Exibir os resultados
print(f"Variância amostral (s^2): {variancia_amostral:.1f} dia^2")
print(f"Desvio padrão amostral (s): {desvio_padrao_amostral:.1f} dia")
````

## **Regra Empírica (68-95-99,7)**

Muitos conjuntos de dados da vida real têm distribuições que são aproximadamente simétricas e em forma de sino.



**Exercício:** Em uma pesquisa conduzida pelo National Center for Health Statistics, a altura média amostral das mulheres nos Estados Unidos (com idades entre 20 e 29 anos) era de 64,2 polegadas, com um desvio padrão amostral de 2,9 polegadas.

Estime a percentagem de mulheres cujas alturas estão entre 58,4 e 64,2 polegada

### **Resolvendo em Python...**

Para estimar a porcentagem de mulheres cujas alturas estão entre 58,4 e 64,2 polegadas em Python, podemos usar a distribuição normal, assumindo que a altura das mulheres segue uma distribuição normal (gaussiana), a mesma utilizada quando calculamos de forma manual.

Vamos calcular a área sob a curva normal que corresponde ao intervalo entre 58,4 e 64,2 polegadas.

**Passos para Resolver o Problema:**
1. Calcular os escores-z para os limites do intervalo (58,4 e 64,2 polegadas).
2. Usar a tabela de distribuição normal padrão (ou a função scipy.stats.norm.cdf) para encontrar a proporção correspondente a esses escores-z.
3. Subtrair os valores para encontrar a porcentagem de mulheres entre esses dois limites.

**Em Python, alguns comandos novos:**

```Python
# Calcular a área sob a curva normal diretamente usando cdf
area1 = norm.cdf(x1, altura_media, desvio_padrao)
area2 = norm.cdf(x2, altura_media, desvio_padrao)
```

**Explicação do Código**

A função `scipy.stats.norm.cdf` permite que você calcule diretamente a área sob a curva normal (ou seja, a probabilidade acumulada) até um determinado valor, simplificando os cálculos que envolvem escores-z e áreas sob a curva.
- O escore-z para 58,4 polegadas (z1) e 64,2 polegadas (z2) é calculado usando a média e o desvio padrão fornecidos.
- `norm.cdf(x, loc=media, scale=desvio_padrao)`: `cdf` retorna a área acumulada sob a curva normal padrão até o valor x. Usamos `norm.cdf` para calcular a probabilidade acumulada até 58,4 polegadas (area1) e até 64,2 polegadas (area2).
- A proporção de mulheres entre 58,4 e 64,2 polegadas é a diferença entre as áreas sob a curva para z2 e z1.
- A proporção é multiplicada por 100 para converter o valor em porcentagem.


````python
# Importa a função 'norm' da biblioteca scipy
from scipy.stats import norm

# Dados
altura_media = 64.2
desvio_padrao = 2.9

# Limites do intervalo
x1 = 58.4
x2 = 64.2

# Calcular a área sob a curva normal diretamente usando cdf
area1 = norm.cdf(x1, altura_media, desvio_padrao)
area2 = norm.cdf(x2, altura_media, desvio_padrao)

# A proporção de mulheres cujas alturas estão entre 58,4 e 64,2 polegadas
proporcao = area2 - area1

# Converter para porcentagem
porcentagem = proporcao * 100

# Exibir o resultado
print(f"A porcentagem estimada de mulheres cujas alturas estão entre 58,4 e 64,2 polegadas é de aproximadamente {porcentagem:.2f}%")
````

## **EXTRA: como criar o gráfico em forma de sino com a área correspondente às alturas**

Para desmonstração, a seguir apresentamos um possível código em Python para criar o gráfico em forma de sino (curva normal ou gaussiana) e visualizar a área correspondente às alturas entre 58,4 e 64,2 polegadas. Aqui, utilizamos as bibliotecas Matplotlib e SciPy.

O gráfico inclui a curva normal padrão e sombreamos a área que representa a porcentagem de mulheres cujas alturas estão entre os dois valores fornecidos.

**Explicação do Código**

- `np.linspace(altura_media - 4*desvio_padrao, altura_media + 4*desvio_padrao, 1000)`: os valores de x são gerados ao redor da média de forma linear de 1000 pontos, abrangendo 4 desvios padrões à esquerda da média (`altura_media - 4*desvio_padrao`) e 4 desvios padrões à direita da média (`altura_media + 4*desvio_padrao`).
- `norm.pdf(x, altura_media, desvio_padrao)`: calcula a densidade de probabilidade para cada valor de x, criando a forma da curva normal.
- `np.linspace(x1, x2, 1000)`: cria uma lista linear de 1000 pontos em `preenchimento_x` com as coordenadas em x que serão sombreadas.
- `norm.pdf(preenchimento_x, altura_media, desvio_padrao)`: cria uma lista em `preenchimento_y` seguindo a função da curva normal para o intervalo entre 58,4 e 64,2 com as coordenadas em y que serão sombreadas.
- `plt.fill_between(preenchimento_x, preenchimento_y, color='skyblue', alpha=0.5, label='Área entre 58.4 e 64.2 polegadas')`: sombreia a área sob a curva na cor `skyblue` entre 58,4 e 64,2 polegadas.
- `plt.axvline(x=altura_media, color='red', linestyle='--', label=f'Média = {altura_media} polegadas')`: adiciona uma linha vertical na média para referência.

````python
# Importa as bibliotecas e funções necessárias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Dados
altura_media = 64.2
desvio_padrao = 2.9

# Limites do intervalo
x1 = 58.4
x2 = 64.2

# Geração de valores para x
x = np.linspace(altura_media - 4*desvio_padrao, altura_media + 4*desvio_padrao, 1000)

# Curva de sino
y = norm.pdf(x, altura_media, desvio_padrao)

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Curva Normal', color='blue')

# Preenchendo a área entre x1 e x2
preenchimento_x = np.linspace(x1, x2, 1000)
preenchimento_y = norm.pdf(preenchimento_x, altura_media, desvio_padrao)
plt.fill_between(preenchimento_x, preenchimento_y, color='skyblue', alpha=0.5, label='Área entre 58.4 e 64.2 polegadas')

# Adicionando legendas e labels
plt.title('Distribuição Normal - Alturas de Mulheres')
plt.xlabel('Altura (polegadas)')
plt.ylabel('Densidade de Probabilidade')
plt.axvline(x=altura_media, color='red', linestyle='--', label=f'Média = {altura_media} polegadas')
plt.legend()

# Mostrando o gráfico
plt.grid(True)
plt.show()
````

## **DEMONSTRATIVO DA TEORIA: Exercício de Desvio padrão para Dados Agrupados**

Uma escola deseja analisar o desempenho dos alunos em uma prova de matemática. As notas foram agrupadas em classes, e a frequência de alunos em cada classe foi registrada conforme a tabela abaixo:

| Nota (Classe) | f (Núm Alunos) |
|---------------|----------------|
| 0-2           | 4              |
| 3-5           | 9              |
| 6-8           | 12             |
| 9-11          | 5              |

````python
# Importa as bibliotecas
import pandas as pd
from math import sqrt

# Dados: frequência de alunos em cada classe
dados = {
    'Nota (Classe)': ['0-2', '3-5', '6-8', '9-11'],
    'f (Núm Alunos)': [4, 9, 12, 5]
}

# Criar DataFrame
df = pd.DataFrame(dados)

# Determinar os pontos médios de cada classe
df['Ponto Médio (x)'] = df['Nota (Classe)'].apply(lambda x: (int(x.split('-')[0]) + int(x.split('-')[1])) / 2)

# Cálculo x*f
df['x*f'] = df['Ponto Médio (x)'] * df['f (Núm Alunos)']

# Cálculo do desvio
df['Desvio (x-u)'] = df['Ponto Médio (x)'] - df['x*f'].sum() / df['f (Núm Alunos)'].sum()

# Calcula os quadrados dos desvios parciais
df['Quadrados (x-u)^2'] = df['Desvio (x-u)']**2

# Calcula os quadrados dos desvios parciais vezes a frequência
df['Quadrados (x-u)^2 * f'] = (df['Desvio (x-u)']**2) * df['f (Núm Alunos)']

# Exibir o DataFrame com as classes, frequências e ponto médio
display(df)


# Soma x*f
soma_xf = df['x*f'].sum()

# Soma f (Núm Alunos)
soma_f = df['f (Núm Alunos)'].sum()

# Média
media = df['x*f'].sum() / df['f (Núm Alunos)'].sum()

# Calcular a soma dos quadrados dos desvios vezes a frequência
soma_quad_freq = df['Quadrados (x-u)^2 * f'].sum()

# Calcular o desvio padrão amostral
desvio_padrao_amostral = sqrt( df['Quadrados (x-u)^2 * f'].sum() / (df['f (Núm Alunos)'].sum() - 1) )

# Exibir os resultados
print(f'Soma x*f: {soma_xf:.1f}')
print(f'Soma f (Núm Alunos): {soma_f:.1f}')
print(f'Média: {media:.1f}\n')

print(f'Quadrados (x-u)^2 * f: {soma_quad_freq:.1f}')
print(f'Desvio padrão amostral (s): {desvio_padrao_amostral:.2f}')
print(f'Soma dos quadrados dos desvios vezes a frequência: {soma_quad_freq:.1f}')
````
